# mkdir Output Tutorial

Notebook conversion of `example_scripts/mkdir_output_tutorial.py`.

This writes a sample CSV to the notebook-local `output/` folder.

## Context Files

The `data/` folder includes real context files copied from `problems/MinimalExample/`:
- `minimal_problem_v3.xml`
- `minimal_model_v1.1.mdl`

```python
from pathlib import Path

import pandas as pd
```

```python
def main():
    # Resolve an output folder path relative to this notebook folder. It may not exist yet.
    output_dir = Path("output")

    # Create the folder tree if needed.
    output_dir.mkdir(parents=True, exist_ok=True)

    # Build an output file path and write a sample table.
    output_csv = output_dir / "sample_results.csv"
    demo_df = pd.DataFrame(
        {
            "Scenario": ["A", "B", "C"],
            "Score": [0.91, 0.87, 0.95],
        }
    )
    demo_df.to_csv(output_csv, index=False)

    print(f"Wrote: {output_csv}")
    return demo_df
```

```python
main()
```
