# BorgRW Example Notebooks

This is the index to all interactive notebook examples demonstrating `borgRWhelper` workflows.

Each example is self-contained with its own data and output folders.

## Nowak Stochastic Simulation

### [Basic Examples](nowak_basic_examples/nowak_basic_examples.ipynb)
Single-year and multi-trace stochastic flow disaggregation using observed monthly proportions.
- Demonstrates `choose_analog_index()` and `sim_multi_trace()`
- Real context: minimal RiverWare model from `problems/MinimalExample/`

### [Oak Creek Example](nowak_oakcreek_example/nowak_oakcreek_example.ipynb)
Real-world workflow with observed Oak Creek monthly data and simulated annual scenarios.
- Reads from Excel; writes simulated traces to Excel output
- Data: `data/oak_creek_monthly_af.xlsx`

## Tradeoff Analysis

### [Sort and Plot Tradeoffs](sort_and_plot_tradeoffs/sort_and_plot_tradeoffs.ipynb)
Load Borg multi-objective solutions and create parallel-axis and scatter-matrix plots.
- **Note:** Expects `AllPolicies.csv` (not currently in repo)
- Includes real context: RedRiver model, problem XML, solution CSV
- Generates interactive HTML visualizations

## Utilities

### [Minimal Dash App](minimal_dash_app/minimal_dash_app.ipynb)
Interactive dashboard using Plotly and Dash with linked dropdown and line charts.
- Demonstrates Dash callbacks and reactive layout
- Uses public Gapminder data

### [mkdir Output Tutorial](mkdir_output_tutorial/mkdir_output_tutorial.ipynb)
Simple example showing how to safely create output directories and write results.
- Uses `Path.mkdir()` for cross-platform robustness
- Demonstrates best practice for output file paths
