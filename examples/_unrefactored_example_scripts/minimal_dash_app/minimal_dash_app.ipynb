# Minimal Dash App

Notebook conversion of `example_scripts/minimal_dash_app.py`.

This app pulls a public dataset from Plotly and creates two linked line charts.

```python
from dash import Dash, html, dcc, callback, Output, Input
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/gapminder_unfiltered.csv")

app = Dash()

app.layout = [
    html.H1(children="Title of Dash App", style={"textAlign": "center"}),
    dcc.Dropdown(df.country.unique(), "Canada", id="dropdown-selection"),
    dcc.Graph(id="pop-graph-content"),
    dcc.Graph(id="life-graph-content")
]


@callback(
    Output("pop-graph-content", "figure"),
    Input("dropdown-selection", "value")
)
def update_pop_graph(value):
    dff = df[df.country == value]
    return px.line(dff, x="year", y="pop")


@callback(
    Output("life-graph-content", "figure"),
    Input("dropdown-selection", "value")
)
def update_life_graph(value):
    dff = df[df.country == value]
    return px.line(dff, x="year", y="lifeExp")
```

## Run the App

Run the next cell to launch the server. Stop the kernel or interrupt execution to end it.

```python
app.run(debug=True)
```
